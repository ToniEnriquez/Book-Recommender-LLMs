```python
from statistics import correlation

import kagglehub
from seaborn import heatmap
from sympy import transpose

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
books = pd.read_csv(f"{path}/books.csv")
```

```python
books
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar=False, ax=ax)

plt.xlabel("Columns")
plt.ylabel("Missing Values")

plt.show()
```

We can see a pattern here: observation that if it has one of them missing (the last 3: average_rating, num_pages and ratings_count), has all of them missing.
Assumptions:
they might be from another dataset?

```python
# take close look at those that have missingness
import numpy as np

books['missing_description'] = np.where(books["description"].isna(), 1 ,0)
books['age_of_books'] = 2024 - books['published_year']
```

```python
books
```

```python
columns_of_interest = ["num_pages", "age_of_books", 'missing_description', 'average_rating']

correlation_matrix = books[columns_of_interest].corr(method="spearman")

sns.set_theme(style="white")
plt.figure(figsize = (8,6))
heatmap = sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f",
                      cbar_kws={"label": "Spearman correlation"})
heatmap.set_title("Correlation Heatmap")
plt.show()
```

we can see that they're not very strong correlated. (close to 1 and -1 is strong, close to 0 is weak).
let's just delete those with no descriptions?

```python
books[(books["description"].isna()) | (books["num_pages"].isna()) | books["average_rating"].isna() | books["published_year"].isna()]
```

since only 303 missing, lets just remove them!

now we make a new dataframe that DO NOT have these columns missing!

```python
books_missing = books[(~books["description"].isna()) & (~books["num_pages"].isna()) & (~books["average_rating"].isna()) & (~books["published_year"].isna())]
```

```python
books_missing #books with none of missing values in those columns above
```

```python
books_missing["categories"].value_counts().reset_index().sort_values("count", ascending=False)
#categories most popular at the top, those not are at the bottom.
# chart settings: bar graph, y-axis:count
```

#Eg. bad one is row 7 -> Description has one word which is "Donation"

```python
books_missing["words_in_description"] = books_missing["description"].str.split().str.len()
```

```python
books_missing
#see the count on y axis
```

```python
#need a cutoff point - how many words in desc in order to be meaningful?
books_missing.loc[books_missing["words_in_description"].between(1, 4), "description"]
```

```python
books_missing.loc[books_missing["words_in_description"].between(5, 14), "description"]
```

```python
books_missing.loc[books_missing["words_in_description"].between(15, 24), "description"]
```

```python
books_missing.loc[books_missing["words_in_description"].between(25, 34), "description"]
#looks like >25 is going to be the cut off!
```

```python
books_missing_25_words = books_missing[books_missing["words_in_description"] >=25 ]
```

```python
books_missing_25_words
#this is the new dataset, now we handle the subtitle
```

```python
books_missing_25_words["title_and_subtitle"] = (
    np.where(books_missing_25_words["subtitle"].isna(), books_missing_25_words["title"],
             books_missing_25_words[["title", "subtitle"]].astype(str).agg(": ".join, axis=1))
)
```

```python
books_missing_25_words
```

```python
books_missing_25_words["tagged_description"] = books_missing_25_words[["isbn13", "description"]].astype(str).agg(" ".join, axis=1)
```

```python
books_missing_25_words
```

```python
# DATASET IS DONE
(
    books_missing_25_words
    .drop(["subtitle", "missing_description", "age_of_books", "words_in_description"], axis=1)
    .to_csv("books_cleaned.csv", index=False)
)
```
